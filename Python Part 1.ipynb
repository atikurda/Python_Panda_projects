```python
first_num = float(input("Enter your first number here to do calculation:"))
second_num = float(input("Enter your second number here to do calculation:"))
result1 = first_num + second_num
result2 = first_num - second_num
result3 = first_num * second_num
result4 = first_num / second_num
print("Your result for addition" , result1)
print("Your result for subtraction" , result2)
print("Your result for multiply" , result3)
print("Your result for division" , result4)
```
